```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
from mpl_toolkits.mplot3d import Axes3D
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
X = iris.data
```

```python
X
```

```python
y = iris.target
```

```python
y
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
X_pca = pca.fit_transform(X)
```

```python
X_pca
```

```python
plt.scatter(X_pca[:,0],X_pca[:,1],c=y,cmap='rainbow')
plt.show()
```

```python
pca = PCA(n_components=3)
```

```python
X_pca3 = pca.fit_transform(X)
```

```python
X_pca3
```

```python
fig = plt.figure(figsize=(8,10))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X_pca3[:,0],X_pca3[:,1],X_pca3[:,2],c=y,cmap='rainbow')
plt.show()
```

```python
# Kernel PCA
```

```python
from sklearn.datasets import make_moons
```

```python
X,y = make_moons(n_samples=500,noise=0.02,random_state=0)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='rainbow')
plt.show()
```

```python
from sklearn.decomposition import KernelPCA
```

```python
kpca = KernelPCA(kernel='rbf',gamma=13)
```

```python
X_kpca = kpca.fit_transform(X)
```

```python
X_kpca
```

```python
plt.scatter(X_kpca[:,0],X_kpca[:,1],c=y,cmap='rainbow')
plt.show()
```
